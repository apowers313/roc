# Basic

```python
sentence = """At eight o'clock on Thursday morning Arthur didn't feel very
           good."""

tokens = nltk.word_tokenize(sentence)
print("tokens:", tokens)

tagged = nltk.pos_tag(tokens)
print("tagged:", tagged)

entities = nltk.chunk.ne_chunk(tagged)
print("entities:", entities)

# from nltk.corpus import treebank
# t = treebank.parsed_sents('wsj_0001.mrg')[0]
# t.draw()
# ModuleNotFoundError: No module named 'tkinter'
```

# Stemming and Lemmatizing

```python
from nltk.stem import PorterStemmer, SnowballStemmer, ARLSTem, LancasterStemmer, WordNetLemmatizer

ps = PorterStemmer()
sbs = SnowballStemmer("english")
arls = ARLSTem()
lan = LancasterStemmer()
wnl = WordNetLemmatizer()

print("\n-- Porter --")
words = ["program", "programs", "programmer", "programming", "programmers"]
for w in words:
    print(f"word: {w}, porter stemmed: {ps.stem(w)}")

print("\n-- Snowball --")
for w in words:
    print(f"word: {w}, snowball stemmed: {sbs.stem(w)}")

print("\n-- Lancaster --")
for w in words:
    print(f"word: {w}, lancaster stemmed: {lan.stem(w)}")

print("\n-- ARLSTem --")
for w in words:
    print(f"word: {w}, arlstem stemmed: {arls.stem(w)}")

print("\n-- WordNetLemmatizer --")
for w in words:
    print(f"""word: {w}, wordnet lemmatized: {wnl.lemmatize(w)}""")
```
